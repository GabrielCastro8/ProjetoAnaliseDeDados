##Quais são as categorias com menor receita nos ultimos 12 meses?

```python
# Unir os dataframes olist_order_items e olist_products
merged_df = pd.merge(T_olist_order_items, T_olist_products, on='product_id')

# Calcular a receita para cada item de pedido
merged_df['total_revenue'] = merged_df['price'] * merged_df['order_item_id']


# Converter order_purchase_timestamp para datetime
T_olist_orders['order_purchase_timestamp'] = pd.to_datetime(T_olist_orders['order_purchase_timestamp'])

# Encontrar a data mais recente na base de dados
last_date_in_dataset = T_olist_orders['order_purchase_timestamp'].max()

# Filtrar os pedidos dos últimos 12 meses
cutoff_date = last_date_in_dataset - pd.DateOffset(months=12)
last_12_months_orders = T_olist_orders[T_olist_orders['order_purchase_timestamp'] >= cutoff_date]

# Unir merged_df com last_12_months_orders
merged_last_12_months = pd.merge(merged_df, last_12_months_orders, on='order_id')

# Agrupar por categoria e somar a receita
revenue_by_category = merged_last_12_months.groupby('product_category_name')['total_revenue'].sum()

# Encontrar as categorias com maior e menor receita
categoria_maior_receita = revenue_by_category.idxmax()
categoria_menor_receita = revenue_by_category.idxmin()

print("Categoria com maior receita nos últimos 12 meses:", categoria_maior_receita)
print("Categoria com menor receita nos últimos 12 meses:", categoria_menor_receita)
```

##Existem sellers que vendem o mesmo produto? Se sim quais são? Qual é a variação de preço praticada entre os sellers?

```python
# Juntar os dataframes olist_order_items e olist_products
merged_order_product = pd.merge(T_olist_order_items, T_olist_products, on='product_id')

# Filtrar as colunas relevantes
merged_order_product = merged_order_product[['seller_id', 'product_id', 'product_category_name', 'price']]

# Agrupar por seller_id e product_id e contar quantas vezes cada combinação ocorre
seller_product_counts = merged_order_product.groupby(['seller_id', 'product_id']).size().reset_index(name='count')

# Filtrar para identificar as combinações de seller_id e product_id que ocorrem mais de uma vez (ou seja, sellers que vendem o mesmo produto)
duplicate_sellers_products = seller_product_counts[seller_product_counts['count'] > 1]

# Juntar os dados duplicados com o dataframe original para obter informações completas sobre os sellers e produtos
duplicate_info = pd.merge(duplicate_sellers_products, merged_order_product, on=['seller_id', 'product_id'])

# Calcular a variação de preço entre os sellers que vendem o mesmo produto
price_variation = duplicate_info.groupby(['seller_id', 'product_id', 'product_category_name'])['price'].agg(['min', 'max'])

# Calcular a variação de preço como a diferença entre o preço máximo e mínimo
price_variation['price_difference'] = price_variation['max'] - price_variation['min']

# Exibir sellers que vendem o mesmo produto e a variação de preço entre eles
print("Sellers que vendem o mesmo produto e variação de preço:")
print(price_variation)
```

Primeiramente observando a relação entre a quantidade de vendas e a quantidade de reviews para os sellers

```python
# Calcular a quantidade de vendas para cada seller
sales_count_by_seller = T_olist_order_items['seller_id'].value_counts().reset_index()
sales_count_by_seller.columns = ['seller_id', 'sales_count']

# Calcular a quantidade de reviews para cada seller
data_merged_reviews_orders = pd.merge(T_olist_order_reviews, T_olist_order_items, on='order_id')
reviews_count_by_seller = data_merged_reviews_orders['seller_id'].value_counts().reset_index()
reviews_count_by_seller.columns = ['seller_id', 'reviews_count']

# Juntar as informações de vendas e reviews em um único dataframe
merged_sales_reviews = pd.merge(sales_count_by_seller, reviews_count_by_seller, on='seller_id', how='inner')

# Calcular a correlação entre as quantidades de vendas e reviews
correlation = merged_sales_reviews['sales_count'].corr(merged_sales_reviews['reviews_count'])

print("A correlação entre Quantidade de Vendas e Quantidade de Reviews é:", correlation)
```

```python
# Plotar o gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_sales_reviews, x='sales_count', y='reviews_count', color='blue')
plt.xlabel('Quantidade de Vendas')
plt.ylabel('Quantidade de Reviews')
plt.title('Relação entre Quantidade de Vendas e Quantidade de Reviews')
plt.grid(True)
plt.tight_layout()
plt.show()
```
